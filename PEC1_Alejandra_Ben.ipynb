```python
#Creamos la clase Person.
```

```python
class Person:
    def __init__(self, first_name, last_name, date_of_birth, job_title):
        self.first_name = first_name
        self.last_name = last_name
        self.date_of_birth = date_of_birth
        self.job_title = job_title

    def get_birth_year(self):
        # Extraer el año de nacimiento como un entero
        return int(self.date_of_birth.split('/')[-1])  # Asume formato dd/mm/yyyy

    def print_info(self):
        # Imprimir la información en el formato solicitado
        birth_year = self.get_birth_year()
        print(f"{self.first_name} {self.last_name} nació en {birth_year}, trabaja como {self.job_title}")


```

```python
# Filtrar personas con la profesión "Broadcast engineer"
broadcast_engineers = df[df['Job Title'] == 'Broadcast engineer']

# Crear una lista para almacenar los objetos Person
person_list = []

# Recorrer el DataFrame filtrado y crear objetos Person
for _, row in broadcast_engineers.iterrows():
    person = Person(
        first_name=row['First Name'],
        last_name=row['Last Name'],
        date_of_birth=row['Date of birth'],
        job_title=row['Job Title']
    )
    person_list.append(person)

# Recorrer la lista de objetos Person y llamar al método print_info
for person in person_list:
    person.print_info()
```

```python
def count_professionals_by_sex(df, profession='Therapeutic radiographer'):
    """
    Función que cuenta el número de profesionales por sexo para una profesión específica.

    Parámetros:
    - dataframe: El DataFrame que contiene los datos.
    - profession: La profesión a filtrar (por defecto 'Therapeutic radiographer').

    Retorna:
    - Un diccionario con el número de profesionales por sexo (Male o Female).
    """
    # Filtrar el DataFrame por la profesión especificada
    filtered_df = df[df['Job Title'] == profession]

    # Contar el número de profesionales por sexo
    sex_counts = filtered_df['Sex'].value_counts()

    # Convertir los resultados a un diccionario
    result = {
        'Male': sex_counts.get('Male', 0),  # Si no hay 'Male', devuelve 0
        'Female': sex_counts.get('Female', 0)  # Si no hay 'Female', devuelve 0
    }

    return result
```

```python
# 1 Validamos los correos electrónicos,usamos una empresión regular que verificael formato nombre@dominio donde el dominio debe contener un punto y al menos un caracter después del punto.
```

```python
import re
```

```python
#Lista de correos electrónicos
emails = [
    "usuario@example.com",
    "otro.usuario@dominio.co",
    "correo.incorrecto@",
    "sinarroba.com",
    "nombre@dominio",
    "usuario@dominio.com",
    "usuario@dominio.c",
    "user.name@sub.dominio.org"
]
```

```python
# Expresión regular para validar correos electrónicos
email_regex = r'^[\w\.-]+@[\w\.-]+\.[a-zA-Z]{2,}$'
```

```python
# Validamos cada correo
for email in emails:
    if re.match(email_regex, email):
        print(f"Válido: {email}")
    else:
        print(f"No válido: {email}")
```

```python
#Explicacion de la expresion regular.
#'^[\w\.-]: el nombre del correo debe de empezar por uno o más caracteres alfanumericos(\),puntos(.)o guiones(-).
#@ debe de contener el símbolo @
#[\w\.-]+ :el dominio debe contener uno o mas caracteres alfanúmericos , puntos o guiones.
#\.[a-zA-Z]{2,}$ . el dmonio debe de terminar con un punto seguido de al menos 2 caracteres alfanumericos ( como .com,.org,.es)
```

```python
# 2.- Extraer las fechas válidas de un texto.
```

```python
# Texto con fechs.
texto = """
Hoy es 12/05/2020 y mañana será 13-05-2020.
En 03.07.2019 se celebró un evento importante.
A veces las fechas vienen en formatos reducidos como 7/8/20 o 07-08-20.
Pero no todas las combinaciones son fechas: 99/99/9999 no es válida.
También pueden aparecer números como 2020/05/12 que no siguen el formato esperado.
"""
```

```python
# Expresión regular para extraer fechas
fecha_regex = r'\b\d{1,2}[-/.]\d{1,2}[-/.]\d{2,4}\b'

# Extraer todas las fechas válidas
fechas = re.findall(fecha_regex, texto)

# Imprimir las fechas encontradas
print("Fechas encontradas:")
for fecha in fechas:
    print(fecha)
```

```python
# Explicación de la expresión regular.
#\b : evita capturar partes de palabras
#\d{1,2}:captura 1 o 2 dígitos para el día.
#[-/.]:permite como separador -,/ o .
#\d{1,2}: captura 1 o 2 dígitos para el mes.
#[-/.]:permite como separador -,/ o .
#\d{2,4}: captura 2 o 4 dígitos para el año.
#\b: evita capturar partes de palabras.
```
